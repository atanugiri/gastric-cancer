```python
import pandas as pd
import os

# Folder path
path = '/Users/atanugiri/.cache/kagglehub/datasets/mahdiehhajian/gene-expression-in-gastric-cancer/versions/1'

# Full file path
file_path = os.path.join(path, 'expression profile(8863 genes).csv')

# Read the CSV
df = pd.read_csv(file_path)
df = df.set_index('Unnamed: 0')
df['mean_expression'] = df.mean(axis=1)
df_sorted = df.sort_values('mean_expression', ascending=False)
```

```python
import pandas as pd

# Full file path
file_path = '/Users/atanugiri/Downloads/gastric cancer data analysis/Hypercholesterolemia_gene_links_updated_final.xlsx'

# Read the Excel file skipping first 2 rows
df_excel = pd.read_excel(file_path, skiprows=2)

# Extract genes from specified locations
genes_part1 = df_excel.iloc[0:27, 1]  # Rows 3-29 -> after skipping 2 rows, it's rows 0:27, column 2 (index 1)
genes_part2 = df_excel.iloc[28:82, 0] # Rows 32-84 -> rows 29:82, column 1 (index 0)
genes_part3 = df_excel.iloc[83:178, 0] # Rows 87-179 -> rows 85:178, column 1 (index 0)

# Concatenate all parts into a single list
chol_genes = pd.concat([genes_part1, genes_part2, genes_part3]).dropna().tolist()

print(chol_genes)
```

```python
# Function to clean gene names
def clean_gene_name(name):
    if pd.isnull(name):
        return None
    return str(name).replace(' ', '').replace('-', '').lower()

# Clean df_sorted index (genes)
df_sorted_clean = df_sorted.copy()
df_sorted_clean.index = df_sorted_clean.index.map(clean_gene_name)

# Clean chol_genes list
chol_genes_clean = [clean_gene_name(gene) for gene in chol_genes]

# Now check overlap
common_genes = set(chol_genes_clean) & set(df_sorted_clean.index)

print(f"✅ {len(common_genes)} genes found in df_sorted.")
print(common_genes)
```

## Read tsv files

```python
import pandas as pd

# File paths
file_paths = [
    'tcga_clinical_data.tsv',
    'tcga_gdc_clinical_data.tsv',
    'tcga_pan_can_atlas_2018_clinical_data.tsv',
    'tcga_pub_clinical_data.tsv'
]

# Initialize empty DataFrame
combined_df = pd.DataFrame()

# Read and collect relevant columns from each file
for path in file_paths:
    df = pd.read_csv(path, sep='\t', usecols=['Sample ID', 'Sex'])
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# Drop duplicates based on 'Sample ID', keeping the first occurrence
combined_df_unique = combined_df.drop_duplicates(subset='Sample ID', keep='first')

# Reset index if needed
combined_df_unique.reset_index(drop=True, inplace=True)

# Show the result
print(combined_df_unique.head())
print(f"✅ Total unique samples: {combined_df_unique.shape[0]}")
```
